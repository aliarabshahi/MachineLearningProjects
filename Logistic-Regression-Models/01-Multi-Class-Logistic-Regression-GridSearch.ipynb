### تعیین متغیر های ورودی و خروجی


```python
X = df.drop('species',axis=1)
y = df['species']
```

### Train | Test Split and Scaling


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```


یادمون هست که برای نرمال کردن داده ها، فقط روی داده های ترین، فیت میشه و سپس روی هر دو ترنسفرم و این کار برای جلو گیری از

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Multi-Class Logistic Regression Model
## پیاده سازی و ترین کردن مدل چند متغیره

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# پارامتر اول بیانگر نوع حل کننده برای رسیدن به بهترین ضرایب در الگوریتم گرادیان دیسنت می باشد
# پارامتر دوم تعیین کننده نوع رفتار با کلاس های گوناگون می باشد مثلا یکی در مقابل بقیه
log_model = LogisticRegression(solver='saga',multi_class="ovr",max_iter=5000)
```

## GridSearchCv for Best Hyper-Parameters

روشی عمومی برای مسائلی با پارامتر های قابل تغییر است به همراه کراس ولیدیشن برای یافتن بهترین آن پارامتر ها می باشد. به این صورت که ابتدا  پارامتر های مسئله را تعریف و سپس آن ها را در قالب یک دیکشنری به تابع تحویل می دهیم

```python
# Penalty Type
penalty = ['l1', 'l2']

# Use logarithmically spaced C values (recommended in official docs)
C = np.logspace(0, 4, 10)
```

```python
grid_model = GridSearchCV(log_model,param_grid={'C':C,'penalty':penalty})
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
grid_model.best_params_
```

### اطمینان از کارکرد صحیح مدل


```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
plot_confusion_matrix(grid_model,scaled_X_test,y_test)
```

```python
# همان جدول بالا به صورت درصدی رو تشکیل می دیم
plot_confusion_matrix(grid_model,scaled_X_test,y_test,normalize='true')
```

```python
print(classification_report(y_test,y_pred))
```

## aoc auc curve
 در مسائل چند متغیره از روش دیگری برای رسم این نمودار ها استفاده می کنیم

```python
from sklearn.metrics import roc_curve, auc
```

```python
def plot_multiclass_roc(clf, X_test, y_test, n_classes, figsize=(5,5)):
    y_score = clf.decision_function(X_test)

    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
plot_multiclass_roc(grid_model, scaled_X_test, y_test, n_classes=3, figsize=(16, 10))
```
