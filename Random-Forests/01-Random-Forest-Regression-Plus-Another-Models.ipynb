# Linear Regression
ابتدا یک مدل خطی ساده را پیاده سازی می کنیم

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_preds = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,lr_preds))
```

به نظر خطای قابل قبولی میاد! برای درک بهتر نحوه کارکرد مدل اول رو رسم می کنیم و متوجه کارکرد افتضاح اون میشیم

```python
signal_range = np.arange(0,100)
```

```python
lr_output = lr_model.predict(signal_range.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='Signal',y='Density',data=df,color='black')
plt.plot(signal_range,lr_output)
```

# Polynomial Regression
مدل چند جمله ای

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## تعریف تابعی به منظور راحتی در پیاده سازی مدل ها
در اینجا به منظور پیاده سازی مدل روی داده های ترین و ارزیابی آن روی داده های تست و در نهایت رسم آن، تابع زیر را تعریف می کنیم. این کار باعث میشه بتونیم به راحتی مدل های مختلفی رو روی داده ها پیاده سازی کنیم و کلی صرفه جویی وقت

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    preds = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    print(f'RMSE : {rmse}')
    
    # Plot results
    signal_range = np.arange(0,100)
    output = model.predict(signal_range.reshape(-1,1))
    
    
    plt.figure(figsize=(12,6),dpi=150)
    sns.scatterplot(x='Signal',y='Density',data=df,color='black')
    plt.plot(signal_range,output)
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```

## تعیریف پایپ لاین

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(2),LinearRegression())
```

```python
run_model(pipe,X_train,y_train,X_test,y_test)
```

## مقایسه مدل قبلی با مدل جدید با توان بیشتر
البته همون طور که می دونین باید به بهترین توان با استفاده از بایاس، واریانس، ترید آف! برسم اما چون هدف اصلی این مثال رندوم فارسته، از انجام این کار صرف نظر می کنیم

```python
pipe = make_pipeline(PolynomialFeatures(10),LinearRegression())
run_model(pipe,X_train,y_train,X_test,y_test)
```
