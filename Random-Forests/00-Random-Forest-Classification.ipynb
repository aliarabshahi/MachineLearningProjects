## فراخوانی کتابخانه ها

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## درباره دیتا

ما پیشتر و در الگوریتم درخت تصمیم این مسئله را حل کردیم، اکنون می خواهیم از الگوریتم جنگل های تصادفی و به منظور رسیدن به خطای قابل قبول تری برای چالش کنونی استفاده کنیم

دیتاست مربوط به گروهی از پنگوئن ها با ویژگی های مختلف است که توسط گورمن ویلیامز! در سال 2014 جمع آوری شده است. هدف از این مسئله پیشبینی گونه پنگوئن ها با توجه به این ویژگی های متفاوت است

در ارتباط به ستون ها اطلاعات تفضیلی به شرح زیر آورده شده است
    * species: penguin species (Chinstrap, Adélie, or Gentoo)
    * culmen_length_mm: culmen length (mm)
    * culmen_depth_mm: culmen depth (mm)
    * flipper_length_mm: flipper length (mm)
    * body_mass_g: body mass (g)
    * island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
    * sex: penguin sex


```python
df = pd.read_csv("../DATA/penguins_size.csv")
```

```python
df = df.dropna()
df.head()
```

## تعیین متغیر های ورودی و خروجی
همان طور که گفته شد ورودی های این الگوریتم باید از جنس اعداد باشند، به همین خاطر با استفاده از تابع گت دامیز! این تغییر شکل را نسبت به داده ها انجام می دهیم

```python
X = pd.get_dummies(df.drop('species',axis=1),drop_first=True)
y = df['species']
```

## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Random Forest Classification
 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
help(RandomForestClassifier)
```

 ##  پیاده سازی مدل با پارامتر های مشخص

```python
# از 10 درخت شروع می کنیم
model = RandomForestClassifier(n_estimators=10,max_features='auto',random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

## ارزیابی مدل ترین شده

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,preds)
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

## بررسی فیچر های مهم تر

به این ترتیب متوجه می شویم کدام یک از متغیر ها نقش و اهمیت بیشتری در تعیین خروجی ها دارند

```python
model.feature_importances_
```

## انتخاب بهینه تعداد درخت ها
همان طور که اشاره شد یکی از راه های انتخاب تعداد مناسب درخت ها رسم تعداد متفاوت در مقابل خطای تولید شده در هر یک و در نهایت انتخاب مقداری که از آن به بعد تغییر ملموسی در مدل انتفاق نیفتاده باشد


```python
test_error = []

for n in range(1,40):
    # Use n random trees
    model = RandomForestClassifier(n_estimators=n,max_features='auto')
    model.fit(X_train,y_train)
    test_preds = model.predict(X_test)
    test_error.append(1-accuracy_score(test_preds,y_test))
 
```

```python
plt.plot(range(1,40),test_error,label='Test Error')
plt.legend()
```

به نظر می رسد انتخاب 5 درخت می تواند گزینه نسبتا مناسبی برای این مسئله باشد

متغیر های مسئله به همراه خروجی را در داده ها مشخص می کنیم

```python
X = df.drop("Class",axis=1)
```

```python
y = df["Class"]
```

## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

## GridSearchCv

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators=[64,100,128,200]
max_features= [2,3,4]
bootstrap = [True,False]
oob_score = [True,False]
```

```python
param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap,
             'oob_score':oob_score}  # Note, oob_score only makes sense when bootstrap=True!
```

```python
rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
plot_confusion_matrix(grid,X_test,y_test)
```
