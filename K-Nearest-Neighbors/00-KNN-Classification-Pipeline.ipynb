## تعیین متغیر های ورودی و خروجی
 
به عنوان اولین قدم سعی می کنیم با کاوش و مصور سازی داده ها، وجود کلاس های متفاوت در دیتا ها را کشف کنیم

```python
X = df.drop('Cancer Present',axis=1)
y = df['Cancer Present']
```

## Train|Test Split and Scaling Data

### یه نکنه ی خیلی مهم
 
**در الگوریتم نزدیک ترین همسایه حتما باید 
<br>
FeatureScaling 
<br>
انجام بشه**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## پیاده سازی و ترین کردن مدل با یک همسایه
 


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(scaled_X_train,y_train)
```

## اطمینان از کارکرد صحیح مدل

```python
y_pred = knn_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

از اونجایی که شاخص های
<br>

<br>
بالا هستند احتمالا بالانس خوبی نسبت به داده ها بر قرار بوده است که در ادامه این را چک می کنیم

```python
print(classification_report(y_test,y_pred))
```

## تعیین تعداد بهینه همسایه ها
## k
تا اینجا از این الگوریتم با یک همسایه به یک خروجی رسیدیم. اما تعداد بهینه همسایه ها در مسائل مختلف می تواند متفاوت باشد! از کجا به  مقدار مناسیی از آن برسیم؟
<br>
پاسخ روش
<br>
**elbow**
<br>
است
<br>
به این صورت که با رسم خطاهای مختلف در همسایه های مختلف و تعیین نقطه ای از نمودار که از آنجا به بعد کاهش ملموسی در مدل رخ نمی دهد، به تعداد مناسبی از همسایه ها می رسیم


### تعریف تابعی به منظور اندازه گیری خطا در همسایه های مختلف
برای این منظور تابعی را تعریف  و نمودار مذکور را رسم می کنیم

```python
test_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train,y_train) 
   
    y_pred_test = knn_model.predict(scaled_X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

**به نظر می رسد که از تعداد همسایه برابر 6، دیگه تغییر ملموسی در کاهش خطا رخ نمیده بنابرین می تونیم این مقدار رو برای مسئله مون در نظر بگیریم**

## یافتن بهترین مقدار برای همسایه بدون در نظر گرفتن پیچیدگی مدل
همیشه با رسم نمودار فوق متوجه می شیم که نقطه ای وجود داره که در اون کمترین خطا یافت می شه. اگر چه با توجه به نظر گرفتن پیچیدگی مدل، **لزوما این بهترین جواب برای ما نیست** اما با روش کراس ولیدیشن و گرید سرچ سعی می کنیم تعداد همسایه برای دریافت کمتری خطا رو هم در این مسئله پیدا کنیم 

### ساخت پایپ لاین
### Pipeline
منظور از پایپ لاین اتوماتیک کردن مجموعه ای از اقدامات، برای راحتی و تسریع در انجام فرایند های مشابه در جاهای دیگر است. یه خورده قلق خاصی داره ایجاد کردنش، اما آسونی که بعدش نصیمبون می شه ارزشش رو داره
<br>

پایپ لاین ما در این مسئله قراره دو کار رو انجام بده:

1. Scale Data
2. Create Model on Scaled Data



```python
scaler = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python
# برای پایپلاین تعریف می شه و توصیه میشه اسامی و متغیر ها هم نام باشن
# ترتیب اهمیت داره و پایپ لاین به همین تربیت اجرا می کنه

operations = [('scaler',scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
```

و به  این صورت پایپ لاین که شامل دو مرحله ی استاندارد سازی داده ها و فیت کردن مدل است رو اجرا می کنه. توجه کنین که سایکیت لرن آنقدر باهوش هست که استاندارد های لازم رو در استاندار سازی داده ها رعایت کنه

```python
pipe = Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

در قدم بعدی می خوایم گرید سرچ سی وی رو روی این پایپ لاینی که ایجاد کردیم، پیاده سازی کنیم. زمانی که پایپ لاین داریم، یه خورده روش پیاده 
سازی پارامتر ها در گرید سرچ سی وی متفاوت تر میشه
<br>

قبلا می گفتیم

 <br>
 اما الان می گیم:

<br>

به این صورت که اول اسم متغیر رو در پایپ لاین که قراره پارامتر های مختلفی روی اون اجرا بشه رو می آوریم
بعد دو تا

```python
k_values = list(range(1,20))
```

```python
k_values
```

```python

param_grid = {'knn__n_neighbors': k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
```

```python
# Use full X and y if you DON'T want a hold-out test set
# Use X_train and y_train if you DO want a holdout test set (X_test,y_test)
full_cv_classifier.fit(X_train,y_train)
```
